## App layout (Diseño de la aplicacion)

```python
app = Dash(__name__)
app.layout = html.Div([

    html.H1("Web Application Dashboards with Dash", style={'text-align': 'center'}), #cabecero h1. Header
    
    #primera mini prueba con un menu desplegable.
    dcc.Dropdown(id="slct_yearsCode",  
                 options=[ #el usuario va a ver las label.
                     {"label": "1", "value": "1"},
                     {"label": "2", "value": "2"},
                     {"label": "3", "value": "3"},
                     {"label": "4", "value": "4"}],
                 multi=False,
                 value=2015,
                 style={'width': "40%"}
                 ),

    html.Div(id='output_container', children=[]),
    html.Br(), #break

    dcc.Graph(id='my_survey_map', figure={})

])
```

```python
app.run_server(debug=True)
```
