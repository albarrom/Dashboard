# Primer Dashboard

```python
# importar librerias

import pandas as pd
import plotly.express as px  
from dash import Dash, dcc, html, Input, Output


#crear un dataframe con toda la informacion de la encuesta
df = pd.read_csv ('../data/survey_results_public2021.csv', index_col = [0]) # El indice sera la columna con el ID de la respuesta
df #mostrar df ()
```

## Tratamiento de datos de las columnas que se van a usar


```python
# Algunas columnas tienen datos que no van a poder procesados correctamente. Por ejemplo:
#less than 1 year y More than 50 years seran problematicos a la hora de hacer estadisticas
df['YearsCode'] = df['YearsCode'].replace(['More than 50 years', 'Less than 1 year'],["50.5", "0.5"])

#convertir todos los NaN del dfen 0
df.fillna("0", inplace=True) 

#YearsCode es ahora una columna de enteros. 
df.YearsCode = df.YearsCode.astype(float)

df
```

## App layout (Diseño de la aplicacion)

```python
app = Dash(__name__)
app.layout = html.Div([

    html.H1("Dashboard ejemplo", style={'text-align': 'center'}), #cabecero h1. Header
    
    #primera mini prueba con un menu desplegable.
    dcc.Dropdown(id="slct_OS",  
                 options=[ #el usuario va a ver las label.
                     {"label": "Windows", "value": "Windows"},
                     {"label": "Linux-based", "value": "Linux-based"},
                     {"label": "MacOS", "value": "MacOS"},
                     {"label": "Windows Subsystem for Linux", "value": "Windows Subsystem for Linux (WSL)"},
                     {"label": "BSD", "value": "BSD"},
                     {"label": "Other", "value": "Other (please specify):"}],
                 multi=False,
                 value="Windows",
                 style={'width': "40%"}
                 ),

    dcc.Graph(id='my_survey', figure={}) # graph container

])
```

# Conectar los gráficos plotly con los componentes dash 

Para crear un dashboard completo, hay que conectar los componentes del app layout usando callback. 

Callback va a usar los ID y los *component property*

```python
@app.callback(
    Output(component_id='my_survey', component_property='figure'),
    Input(component_id='slct_OS', component_property='value'))
def update_graph(option_slctd):
    #filtered_df = df[df.year == selected_year]
    
    dff = df.copy() #copia del df
    dff = dff[dff["OpSys"] == option_slctd]
        
    fig = px.line(dff, x= dff['YearsCode'].sort_values(), 
                  y=dff['Age'].sort_values(), 
                  color=dff['EdLevel']) 
    return fig
```

```python
app.run_server(debug=True, use_reloader=False)
```
