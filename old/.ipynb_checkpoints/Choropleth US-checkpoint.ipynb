# [Choropleth](https://plotly.com/python/mapbox-county-choropleth/) + tabs

## 1. Importar las librerías + el csv con los datos de la encuesta. 

```python
# importar librerias

import pandas as pd
import plotly.express as px  
from dash import Dash, dcc, html, Input, Output
import us

#crear un dataframe con toda la informacion de la encuesta
df_csv = pd.read_csv ('../data/survey_results_public2021.csv', index_col = [0]) # El indice sera la columna con el ID de la respuesta
df_csv #mostrar df ()
```

## 2. Preprocesar datos. 

Tratar las columnas/conjunto de datos para comenzar a crear los gráficos. 


Primer dashboard: total de personas que han realizado la encuesta por Estado. (Solamente se tiene en cuenta US de momento)

```python

df_states = df_csv.copy() #copia del df
#eliminar todos los valores que no sean relevantes 
df_states = df_states.drop(df_states[df_states.US_State == "I do not reside in the United States"].index)
df_states = df_states.dropna(subset=['US_State'])

#se usa una libreria para crear una nueva columna con las abreviaturas de los estados
abreviaturas =[]
for i in df_states['US_State']: abreviaturas.append(us.states.lookup(i).abbr)
df_states['Short_US'] = abreviaturas

#numero de personas que han realizado la encuensta por estado
df_states['total']=1
df_states['Counts'] = df_states.groupby(['US_State'])['total'].transform('count')
df_states.drop('total', axis=1, inplace=True) 

df_states # nueva columna con la abreviatura de los estados. 
```

## 3. Gráfico.

En este caso, un mapa de US

```python
app = Dash(__name__)
# App layout
app.layout = html.Div([
    dcc.Tabs(id='tabs-example-1', value='tab-1', 
             children=[
                dcc.Tab(label='US', value='tab-1'),
                dcc.Tab(label='Resto del mundo', value='tab-2'), #por hacer
    ]),
    html.Div([
        html.Div(id='tabs-example-content-1')
    ])
])
```

```python
#app callback

@app.callback(
    
    Output('tabs-example-content-1', 'children'),
    Input('tabs-example-1', 'value')
)

def render_content(tab):
    if tab == 'tab-1':
        return html.Div([
            html.H3('Tab content 1'),
            dcc.Graph( #grafico de US
                figure=px.choropleth(
        data_frame=df_states,
        locationmode='USA-states',
        locations='Short_US',
        scope="usa",
        color='Counts',
        hover_data=['US_State', 'Counts'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        labels={'Counts': 'numero de encuestados', 'Short_US':'Estado USA'}
        
    )
            )
        ])
    elif tab == 'tab-2': #Por hacer
        return html.Div([
            html.H3('Tab content 2'),
            dcc.Graph(
                figure=dict( #grafico de ejemplo
                    data=[dict(
                        x=[1, 2, 3],
                        y=[5, 10, 6],
                        type='bar'
                    )]
                )
            )
        ])

```

```python
app.run_server(debug=True, use_reloader=False)
```
