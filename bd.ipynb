```python
# importar librerias

import sqlite3
import pandas as pd

#crear un dataframe con toda la informacion de la encuesta
df = pd.read_csv ('survey_results_public.csv', index_col = [0]) # El indice sera la columna con el ID de la respuesta
df #mostrar df ()
```

```python
# cargar los datos del csv en un dataframe
df = pd.read_csv('survey_results_public.csv')

# se eliminan los espacios en blanco de la cabecera
# csv.columns = df.columns.str.strip()

# Crear un objeto que conecte a la bd
con = sqlite3.connect("survey.db")

# escribir los datos en una db de sqlite
df.to_sql('tabla2021', con, if_exists='replace', index = False)

# Una vez creada la conexion, crear un objeto cursor que nos permitira ejecutar los comandos en SQL.
cursor = con.cursor()

#Mostrar resultado
cursor.execute('''SELECT * FROM tabla2021''').fetchall()
#for row in cursor.execute('SELECT * FROM csv'):
  #  print(row)
```

```python
#opcionalmente, la consulta tambien se puede mostrar como un df
pd.read_sql('''SELECT * FROM tabla2021''', con)
```

```python
# Test de algunos ejemplos de consultas:

print ("Numero de personas nativas de Canada y Eslovaquia")
print(cursor.execute('''SELECT Country, count(*) AS Ciudad 
                  FROM tabla2021 where Country in('Canada','Slovakia') 
                  group by Country''').fetchall())

#problematico. no todos los valores en YearsCode son enteros.
print ("\nNumero de personas que han estado desarrollando mas de 7 años")
print(cursor.execute('''SELECT COUNT(*) FROM tabla2021 
                        WHERE YearsCode > 7''').fetchall())
```

```python
#Ejemplo mostrando directamente el df

print("Conteo del tipo de educacion de los encuestados")
pd.read_sql('''SELECT EdLevel AS 'Nivel de educacion', 
               COUNT(EdLevel) AS 'Total' 
               FROM tabla2021 
               GROUP BY EdLevel''', con) 
```

```python
# cerrar la conexion a la base de datos de sqlite
con.close()
```
